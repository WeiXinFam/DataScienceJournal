```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_m=pd.read_csv('data/movies.csv')
df_m.head()
```

```python
df_l=pd.read_csv('data/links.csv')
df_l.head()
```

```python
df_r=pd.read_csv('data/ratings.csv')
df_r.head()
```

```python
df_m1=pd.merge(df_m,df_l,on='movieId')
df_m1

df_m2=pd.merge(df_m1,df_r,on='movieId')
df_m2.info()
#have null values
```

```python
df_t=pd.read_csv('data/tags.csv')
df_t.head()
```

```python
df_m3=pd.merge(df_m2,df_t,on='movieId')
df_m3.info()
```

```python
df_m3.describe()
```

use the Pearson correlation coefficient. This number will lie between -1 and 1. 1 indicates a positive linear correlation while -1 indicates a negative correlation. 0 indicates no linear correlation. Therefore movies with a zero correlation are not similar at all.
